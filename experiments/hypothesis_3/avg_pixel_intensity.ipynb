```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mnist = pd.read_csv("../datasets/mnist.csv")
mnist.describe()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5), sharey=True)
```

```python
sevens = mnist[mnist.label == 7]
sevens = sevens.drop("label", axis=1)
sevens /= len(sevens)
sum_seven = np.zeros((28, 28))
for idx, seven in sevens.iterrows():
    sum_seven += np.reshape(seven.values, (28, 28))
ax[1].imshow(sum_seven, vmin=0, vmax=200)
ax[1].set_title("Class 7")
ax[1]
```

```python
nums = mnist.drop("label", axis=1)
nums /= len(nums)
sum_nums = np.zeros((28, 28))
for idx, num in nums.iterrows():
    sum_nums += np.reshape(num.values, (28, 28))
im = ax[0].imshow(sum_nums, vmin=0, vmax=200)
ax[0].set_title("All Images")
ax[0]
```

```python
fig.suptitle("Average Pixel Intensity of MNIST Dataset")
fig.colorbar(im, ax=ax.ravel().tolist())
fig
```
